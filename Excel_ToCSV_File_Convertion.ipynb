```python

import numpy as np 
import pandas as pd 
import os


def excel_to_csv(excel_file, csv_file):
    try:
        df = pd.read_excel(excel_file)
        df.to_csv(csv_file, index=False)
        
    except Exception as e:
        print(f"Error occurred while processing {excel_file}: {e}")

excel_dir = "/kaggle/input/2021-olympics-in-tokyo"
csv_dir = "/kaggle/working/"

os.makedirs(csv_dir, exist_ok=True)

for filename in os.listdir(excel_dir):
    if filename.endswith(".xlsx") or filename.endswith(".xls"):
        excel_file = os.path.join(excel_dir, filename)
        csv_file = os.path.join(csv_dir, os.path.splitext(filename)[0] + ".csv")
        excel_to_csv(excel_file, csv_file)
```
